### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
print("Column Names:\n", df.columns)
print()
print("Data Types:\n", df.dtypes)
print()
print("Summary Information:\n", df.describe())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df.isnull().sum()

## Must impute or drop values as necessary for Employment, JobSat, YearsCodePro, LanguageHaveWorkedWith, LanguageWantToWorkWith, and EdLevel

## JobSat and YearsCodePro are numerical
df['JobSat'].fillna(df['JobSat'].median(), inplace=True)
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
df['YearsCodePro'].fillna(df['YearsCodePro'].median(), inplace=True)

## Employment, LanguageHaveWorkedWith, LanguageWantToWorkWith, and EdLevel are categorical
df['Employment'].fillna(df['Employment'].mode()[0], inplace=True)
df['LanguageHaveWorkedWith'].fillna('', inplace=True)
df['LanguageWantToWorkWith'].fillna('', inplace=True)
df['EdLevel'].fillna(df['EdLevel'].mode()[0], inplace=True)

## Check for missing values
print("Missing Values Count:\n", df[['Employment', 'JobSat', 'YearsCodePro', 'LanguageHaveWorkedWith', 'LanguageWantToWorkWith', 'EdLevel']].isnull().sum())
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
print(df['Employment'].value_counts(), '\n\n')
print(df['JobSat'].value_counts(), '\n\n')
print(df['YearsCodePro'].value_counts(), '\n\n')
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
plt.figure(figsize=(10,6))
sns.kdeplot(df['JobSatPoints_1'], fill=True, color='blue')
plt.title('KDE Distribution of Job Satisfaction Scores')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Density')
plt.tight_layout()
plt.show()

## The plot curve shows that most people are deeply unsatisfied with their jobs, with the peak job satisfaction score at around 0-5.
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
df['HaveSet'] = df['LanguageHaveWorkedWith'].fillna('').apply(lambda x: set(x.split(';')) if x else set())
df['WantSet'] = df['LanguageWantToWorkWith'].fillna('').apply(lambda x: set(x.split(';')) if x else set())

have_all = set().union(*df['HaveSet'])
want_all = set().union(*df['WantSet'])

from collections import Counter

have_counts = Counter(lang for langs in df['HaveSet'] for lang in langs if lang)
want_counts = Counter(lang for langs in df['WantSet'] for lang in langs if lang)

all_langs = sorted(set(have_counts.keys()) | set(want_counts.keys()))

plot_df = pd.DataFrame({
    'Language': all_langs,
    'HaveWorkedWith': [have_counts.get(lang, 0) for lang in all_langs],
    'WantToWorkWith': [want_counts.get(lang, 0) for lang in all_langs]
})

plt.figure(figsize=(14,8))
plot_df_melted = plot_df.melt(id_vars='Language', var_name='Category', value_name='Count')
sns.barplot(data=plot_df_melted, x='Language', y='Count', hue='Category', palette='Blues')
plt.xticks(rotation=90)
plt.title('Comparison of Programming Languages: Used vs Desired')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
ct = pd.crosstab(df['Country'], df['RemoteWork'])

plt.figure(figsize=(18,20))
sns.heatmap(ct, cmap='Blues', linewidths=.5)
plt.title('Heatmap of Remote Work Distribution by Country')
plt.xlabel('Remote Work Type')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
## Spearman correlation coefficient is appropriate due to the non-linear data of both JobSat and YearsCodePro

spearman_corr = df[['JobSat', 'YearsCodePro']].corr(method='spearman')
print("Spearman correlation:")
print(spearman_corr)
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
top_emp = df['Employment'].value_counts().nlargest(6).index
df['Employment_Grouped'] = df['Employment'].apply(
    lambda x: x if x in top_emp else 'Other'
)

ct_pct = pd.crosstab(
    df['EdLevel_Grouped'],
    df['Employment_Grouped'],
    normalize='index'
) * 100

ct_pct.plot(
    kind='bar',
    stacked=True,
    figsize=(14,8),
    colormap='Blues'
)

plt.title('Employment Status by Education Level (Top Categories + Other)')
plt.xlabel('Education Level')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
